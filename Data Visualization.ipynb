# Exploratory Data Analysis.
Analysing the Titanic train data set to summarize its main characteristics by providing meaningful insights using Visualization

<h1> Collection of Data </h1>

```python
import pandas as pd
train_data = pd.read_csv('D:\Data Visualization\\Data-Visualization\\train.csv')
train_data.head()
```

<h1> Statistical Summary of the dataset </h1>

```python
train_data.describe()
```

<h1> Cleaning the Data </h1>

```python
train_data.drop('Cabin', axis=1, inplace=True)
```

```python
train_data.head()
```

```python
# used mean to fill empty age values in the dataset
train_data['Age'].fillna(train_data['Age'].mean(), inplace=True)
```

```python
train_data.head()
```

```python
train_data.dropna(inplace=True)
```

```python
train_data.head()
```

```python
# Check whether all missing values are filled
train_data.isnull().sum()
```

<h1> Visualizing the Data </h1>

```python
import seaborn as sns
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Pclass', data=train_data)
```

```python
sns.boxplot(x='Pclass', y='Fare', data=train_data)
```

```python
sns.distplot(train_data['Pclass'], kde=False)
```

```python
sns.catplot(x='Embarked', y='Survived', hue='Sex', col='Pclass', kind='bar', data=train_data)
```

```python
sns.violinplot(x="Sex", y="Survived",data= train_data)
```

```python
sns.pairplot(train_data, hue='Sex')
```

```python
sns.jointplot(x="Age", y="Fare", data=train_data, color ='blue')
```
